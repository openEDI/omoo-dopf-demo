```python
directory = "outputs"
```

```python
with open(os.path.join(directory, "topology.json")) as f:
    topology = Topology.parse_obj(json.load(f))
    base_voltages = np.array(topology.base_voltage_magnitudes.values)
```

```python
voltage_real = feather.read_feather(os.path.join(directory, "voltage_real.feather"))
voltage_imag = feather.read_feather(os.path.join(directory, "voltage_imag.feather"))

time = voltage_real["time"]
true_voltages = voltage_real.drop('time', axis=1) + 1j * voltage_imag.drop('time', axis=1)
true_voltages = (np.abs(true_voltages) / base_voltages).drop(index = [len(true_voltages) -1], axis=0)
true_voltages = true_voltages.values
num_T, num_node = true_voltages.shape

voltage_new_real = feather.read_feather(os.path.join(directory, "voltage_new_real.feather"))
voltage_new_imag = feather.read_feather(os.path.join(directory, "voltage_new_imag.feather"))

new_voltages = voltage_new_real.drop('time', axis=1) + 1j * voltage_new_imag.drop('time', axis=1)
new_voltages = (np.abs(new_voltages) / base_voltages).drop(index = [0], axis=0)
new_voltages = new_voltages.values
```

```python
pv_P_old = feather.read_feather(os.path.join(directory, "pv_P_old.feather")).drop('time', axis=1)
pv_Q_old = feather.read_feather(os.path.join(directory, "pv_Q_old.feather")).drop('time', axis=1)
pv_P_new = feather.read_feather(os.path.join(directory, "pv_P_new.feather")).drop('time', axis=1)
pv_Q_new = feather.read_feather(os.path.join(directory, "pv_Q_new.feather")).drop('time', axis=1)

pv_P_old = pv_P_old.drop(index = [len(pv_P_old) -1], axis=0)
pv_Q_old = pv_Q_old.drop(index = [len(pv_Q_old) -1], axis=0)
pv_P_new = pv_P_new.drop(index = [0], axis=0)
pv_Q_new = pv_Q_new.drop(index = [0], axis=0)
```

```python
def plot_power_changed(t, pv_P_new, pv_Q_new, pv_P_old, pv_Q_old, plot_directory="."):
    all_pv_names = np.array(list(pv_Q_old.columns))
    p_set = pv_P_new.iloc[t,:].values
    q_set = pv_Q_new.iloc[t,:].values
    p_old = pv_P_old.iloc[t,:].values
    q_old = pv_Q_old.iloc[t,:].values
    
    p_fig = plt.figure(figsize=(12,8))
    ax = p_fig.subplots(1)
    sns.lineplot(x=all_pv_names, y=p_old, marker="o", label = 'Old setpoints', ax=ax)
    sns.lineplot(x=all_pv_names, y=p_set, marker="o", label = 'Implemented setpoints', ax=ax)
    # plt.xlabel("PV names")
    ax.legend(fontsize=18)
    ax.set_ylabel("Real part of PV power (kW)")
    ax.set_title("The comparison for active power at time step " + str(t+1), fontsize=18)
    p_fig.savefig(os.path.join(plot_directory, f'Change_P_{t}.png'), dpi=300)

    q_fig = plt.figure(figsize=(12,8))
    ax = q_fig.subplots(1)
    sns.lineplot(x=all_pv_names, y=q_old, marker="o", label = 'Old setpoints', ax=ax)
    sns.lineplot(x=all_pv_names, y=q_set, marker="o", label = 'Implemented setpoints', ax=ax)

    ax.legend(fontsize=18)
    ax.set_ylabel("Imaginary part of PV power (kVar)")
    ax.set_title("The comparison for reactive power at time step " + str(t+1), fontsize=18)
    q_fig.savefig(os.path.join(plot_directory, f'Change_Q_{t}.png'), dpi=300)

    return p_fig, q_fig
```

```python
def plot_voltages_compare(t, true_voltages, new_voltages, Vmin, Vmax, plot_directory = "."):
    Vold, Vnew = true_voltages[t], new_voltages[t]
    fig = plt.figure(figsize=(12,8))
    ax = fig.subplots(1)
    sns.lineplot(x=range(num_node), y=Vold, marker="o", label = 'Voltages without the algorithm', ax=ax)
    sns.lineplot(x=range(num_node), y=Vnew, marker="o", label = 'Voltages with the algorithm', ax=ax)
    sns.lineplot(x=range(num_node), y=Vmax, color='red', linestyle='--', label = 'Limits')
    sns.lineplot(x=range(num_node), y=Vmin, color='red', linestyle='--', ax=ax)
    ax.set_title("The comparison for voltages at time step " + str(t+1), fontsize=18)
    ax.set_xlabel("Time step")
    ax.set_ylabel("Voltages (p.u.)")
    fig.savefig(os.path.join(plot_directory, f'Compare_V_{t}.png'), dpi=300)
```

```python
plot_voltages_compare(73, true_voltages, new_voltages, Vmin, Vmax, plot_directory=plot_directory)
```

```python
plot_power_changed(73, pv_P_new, pv_Q_new, pv_P_old, pv_Q_old, plot_directory=plot_directory)
```

```python
target_time = [0, 32, 38, 46, 51]
V_max, V_min = 1.05, 0.945
plot_directory = os.path.join(directory, 'plots/')
if not os.path.exists(plot_directory):
    os.makedirs(plot_directory)

Vmax = np.ones(num_node) * V_max
Vmin = np.ones(num_node) * V_min
for t in [0, 32, 38, 46, 51]:
    plot_power_changed(t, pv_P_new, pv_Q_new, pv_P_old, pv_Q_old, plot_directory=plot_directory)
    plot_voltages_compare(t, true_voltages, new_voltages, Vmin, Vmax, plot_directory=plot_directory)
print(np.min(new_voltages), np.max(new_voltages))
print(np.sum(pv_P_new, axis =1))
```
